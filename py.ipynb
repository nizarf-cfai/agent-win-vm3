```python
import os
import json
```

```python
a = ['1','f','t','s','t']

for i in a[:-1]:
    print(i)
```

```python
log_list = os.listdir("log_onetime")
for log_file in log_list:
    with open(f"log_onetime/{log_file}", "r", encoding="utf-8") as f:
        data = json.load(f)
    print(log_file, ":", data.keys())
```

```python
log_list = os.listdir("log_error")
for log_file in log_list:
    with open(f"log_error/{log_file}", "r", encoding="utf-8") as f:
        data = json.load(f)
    print(log_file, ":", data.keys())
```
